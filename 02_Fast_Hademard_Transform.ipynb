# Brza Hadamardova transofrmacija

```python
def fwht(l):
    h = 1
    while h < len(l):
        for i in range(0, len(l), h * 2):
            for j in range(i, i + h):
                x = l[j]
                y = l[j + h]
                l[j] = x + y
                l[j + h] = x - y
                
        l = [i / 2 for i in l]
        h *= 2
```

```python
from datetime import datetime
```

```python
l = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
t1 = datetime.now()
fwht(l)
print(l)
fwht(l)
print(l)
t2 = datetime.now()
print(t2 - t1)
```
